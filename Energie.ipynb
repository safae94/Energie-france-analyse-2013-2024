#5 - Création de 3 DataFrames : `df_clean` contient les colonnes énergétiques principales et les informations temporelles ; `df_tco` contient les colonnes TCO/TCH  et TCO nucluaire ; `df_nucleaire` contient `TCH Nucléaire (%)`

## 5.1 creation df_clean

- on remplace zero les   "Nucléaire (MW)",Stockage batterie",Déstockage batterie",Eolien terrestre",Eolien offshore"

```python
df_clean = df_definitive.copy()
df_sansmanquantes = ["Nucléaire (MW)",
    "Stockage batterie",
    "Déstockage batterie",
    "Eolien terrestre",
    "Eolien offshore",]
df_clean[df_sansmanquantes] = df_clean[df_sansmanquantes].fillna(0)
```

- on supprume les valeur manquantes de  "Consommation (MW)","Thermique (MW)","Solaire (MW)","Hydraulique (MW)","Bioénergies (MW)","Ech. physiques (MW)","Pompage (MW)"

```python
df_sup = [
    "Consommation (MW)", "Thermique (MW)", "Solaire (MW)",
    "Hydraulique (MW)", "Bioénergies (MW)",
    "Ech. physiques (MW)", "Pompage (MW)"
]
df_clean = df_clean.dropna(subset=df_sup)
```

```python
df_clean['Eolien (MW)'] =  df_clean['Eolien (MW)'].replace('ND', np.nan).astype(float)
```

```python
df_clean = df_clean.drop(
    columns=df_clean.filter(regex='^(TCH|TCO)').columns.tolist() + ['Column 30']
)
```

## 5.3  creation data df_nucleaire : contient  colonne tch nucluaire

```python

## Colonnes nucléaires
cols_nucleaire = [ 'TCH Nucléaire (%)']

# Régions concernées
regions_2020 = [
    'Auvergne-Rhône-Alpes',
    'Centre-Val de Loire',
    'Grand Est',
    'Hauts-de-France',
    'Normandie',
    'Nouvelle-Aquitaine',
    'Occitanie'
]

df_nucleaire = pd.DataFrame()



df_nucleaire['TCH Nucléaire (%)'] = df_definitive.loc[
    (df_definitive['Région'].isin(regions_2020)) &
    (df_definitive['annee'] > 2019 ),
    'TCH Nucléaire (%)'
]
```

```python

## Colonnes nucléaires
cols_nucleaire = [ 'TCH Nucléaire (%)','TC0 Nucléaire (%)','Région','annee']

# Régions concernées
regions_2020 = [
    'Auvergne-Rhône-Alpes',
    'Centre-Val de Loire',
    'Grand Est',
    'Hauts-de-France',
    'Normandie',
    'Nouvelle-Aquitaine',
    'Occitanie'
]



df_nucleaire = df_definitive[
    (
        (
            df_definitive['Région'].isin(regions_2020)
        ) &
        (
            df_definitive['annee'] > 2020
        )
    )
    |
    (
        (
            df_definitive['Région'].isin(regions_2020)
        ) &
        (
            df_definitive['annee'] > 2019
        )
    )
][['Nucléaire (MW)','TCO Nucléaire (%)', 'TCH Nucléaire (%)','Région','annee']]
```

# 6 - Visualisations et Statistiques

## 6.1 Proportion des données certifiées sur la période étudiée (2013-2026)

```python
x = df['Nature'].value_counts()
# Camembert
plt.figure()
plt.pie(x.values,labels=x.index,autopct='%1.1f%%')
plt.title("Proportion des données certifiées sur la période étudiée (2013-2026)")
plt.axis('equal')
plt.show()
```

## 6.2 Consommation par région

```python
cons_anu = df_clean.groupby(['Région'])['Consommation (MW)'].mean().sort_values()
fig, ax = plt.subplots(figsize=(10, 6))
cons_anu.plot(
    kind="barh",
    ax=ax,
    color="steelblue"
)
```

- On observe une forte disparité entre les régions en termes de production énergétique. L’Île-de-France et Auvergne-Rhône-Alpes présentent les niveaux les plus élevés, tandis que des régions comme Centre-Val de Loire ou Bourgogne-Franche-Comté affichent des valeurs plus faibles. Cela met en évidence une concentration de la production dans certaines régions

## 6.3	Mix énergétique moyen de la France (répartition par source en MW)

- Le mix énergétique français est largement dominé par le nucléaire, qui représente environ 69 % de la production totale. L’hydraulique occupe la deuxième place (environ 17 %), suivi du thermique (8 %) et du solaire (3 %). Les autres sources, comme l’éolien et les bioénergies, restent marginales. Cela montre une forte dépendance au nucléaire dans la production d’électricité en France

```python

mix = df_clean[[
    "Nucléaire (MW)",
    "Thermique (MW)",
    "Solaire (MW)",
    "Hydraulique (MW)",
    "Eolien terrestre",
    "Eolien offshore",
    "Bioénergies (MW)"
]].mean()
fig, ax = plt.subplots(figsize=(8,8))

ax.pie(
    mix.values,
    labels=mix.index,
    autopct='%1.1f%%',
    startangle=180
)

ax.set_title("Mix énergétique moyen de la France")

plt.show()
```

## 6.4 Production moyenne par source


```python
mix = df_clean[['Nucléaire (MW)', 'Thermique (MW)', 'Eolien (MW)',
          'Solaire (MW)', 'Hydraulique (MW)']].mean().sort_values()
plt.figure()
mix.plot(kind='bar')
plt.title("Production moyenne par source")
plt.ylabel("MW")
plt.show()
```

- Le graphique montre que le nucléaire représente la principale source de production d’électricité en France, avec une production moyenne largement supérieure aux autres énergies. L’hydraulique occupe la deuxième place, tandis que l’éolien, le thermique et le solaire contribuent de manière plus modérée à la production électrique nationale

## 6.5 Consommation vs Production par région

```python


df_clean['mix'] = df_clean[['Nucléaire (MW)', 'Thermique (MW)','Solaire (MW)','Hydraulique (MW)','Eolien (MW)',]].sum(axis=1)
region_balance = df_clean.groupby('Région')[['Consommation (MW)', 'mix']].sum()
region_balance.plot(kind='bar')
plt.title("Consommation vs Production par région")
plt.ylabel("MW")
plt.xticks(rotation=90)
plt.show()
```

- Le graphique compare la consommation et la production d’électricité par région. Certaines régions, comme l’Auvergne-Rhône-Alpes et le Grand Est, produisent plus d’électricité qu’elles n’en consomment, tandis que d’autres, comme l’Île-de-France et la Bretagne, consomment davantage qu’elles ne produisent. Cela met en évidence les déséquilibres régionaux et l’importance du réseau électrique national.

## 6.6 Distribution des énergies renouvelables

```python

# Somme des énergies renouvelables
renouvelables = {

    'Solaire': df_clean['Solaire (MW)'].sum(),
    'Eolien (MW)': df_clean['Eolien (MW)'].sum(),
    'Hydraulique': df_clean['Hydraulique (MW)'].sum()

}
# Camembert
plt.figure()
plt.pie(renouvelables.values(), labels=renouvelables.keys(), autopct='%1.1f%%')
plt.title("Distribution des énergies renouvelables")
plt.axis('equal')
plt.show()
```
